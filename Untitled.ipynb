```python
font_scale=3
font=cv2.FONT_HERSHEY_PLAIN
for classInd,conf,boxes in zip(classIndex.flatten(),confidece.flatten(),bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classLabels[classInd-1],(boxes[0]+10,boxes[1]+40),font,fontScale=font_scale,color=(0,255,0),thickness=3)
```

```python
import cv2

cap = cv2.VideoCapture('11_2018_SM_fitness_8_P5.mov')
if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Fallback to webcam if video file cannot be opened
if not cap.isOpened():
    raise IOError("Cannot open video")

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
new_width = 3840
new_height = 2160

while True:
    ret, frame = cap.read()
    if not ret:  # Check if the frame was successfully read
        print("Failed to capture frame")
        break

    # Ensure frame is not empty before detection
    if frame is None or frame.size == 0:
        print("Empty frame, skipping detection")
        continue
    frame_resized = cv2.resize(frame, (new_width, new_height))

    # Assuming 'model' and 'classLabels' are already defined
    classIndex, confidence, bbox = model.detect(frame_resized, confThreshold=0.55)
    print(classIndex)

    if len(classIndex) != 0:
        for classInd, conf, boxes in zip(classIndex.flatten(), confidence.flatten(), bbox):
            if classInd <= 80:
                x, y, w, h = boxes  # Unpack bounding box coordinates
                x = int(x * (new_width / frame.shape[1]))
                y = int(y * (new_height / frame.shape[0]))
                w = int(w * (new_width / frame.shape[1]))
                h = int(h * (new_height / frame.shape[0]))
                cv2.rectangle(frame_resized, (x, y), (x + w, y + h), (255, 0, 0), 2)
                cv2.putText(frame_resized, classLabels[classInd-1], (x + 10, y + 40), font,
                            fontScale=font_scale, color=(0, 255, 0), thickness=3)

    cv2.imshow('Object Detection Tutorial',frame_resized)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
    
```
